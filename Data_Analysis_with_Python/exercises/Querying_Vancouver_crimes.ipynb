# Querying Vancouver crimes

Using the given sqlite3 connection:
- Select `TYPE`, `MONTH`, `DAY` and `NEIGHBOURHOOD` from the `van_crimes` table, but only the crime observations from `Stanley Park` or `West End`. Store the information in a `van_crimes_df` DataFrame.
- Store the count of crimes per `TYPE` in a `crime_types_count` variable.

```python
import sqlite3
import pandas as pd
```

```python

conn= sqlite3.connect(   'van_crime_2003 (1).db'           )#connect to the datbase
cur= conn.cursor()#create a cursor

cur.executescript(open('van_crime_2003 (1).sql', 'r').read())#execute the script in the file


van_crimes_df = pd.read_sql('SELECT * FROM van_crimes', conn)# get and transform in a df
```

```python
query="""SELECT TYPE, MONTH, DAY ,NEIGHBOURHOOD FROM van_crimes where NEIGHBOURHOOD = 'Stanley Park' OR NEIGHBOURHOOD = 'West End' """

location_crime = pd.read_sql(query,conn)

location_crime
```

```python
crime_types_count = van_crimes_df.loc[:,'TYPE'].value_counts()
# muestra tabla type y su cuenta
#agrupados type(se agrupan aeso)
#y organiza count
crime_types_count_query=""" SELECT TYPE, COUNT(*) AS count
FROM van_crimes
GROUP BY TYPE
ORDER BY count DESC;   """
crime_types_count_sql=pd.read_sql(crime_types_count_query,conn)

crime_types_count
```

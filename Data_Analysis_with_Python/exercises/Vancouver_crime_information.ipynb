Using the given sqlite3 connection:
- Store all the crimes committed after 18:00 h in a `late_crimes` variable.
- Store the number of crimes committed on the month with most crimes in a `dangerous_month_crimes` variable.

```python
import sqlite3
import pandas as pd

```

```python
conn= sqlite3.connect(   'van_crime_2003.db'           )#connect to the datbase
cur= conn.cursor()#create a cursor

cur.executescript(open('van_crime_2003.sql', 'r').read())#execute the script in the file


van_crimes_df = pd.read_sql('SELECT * FROM van_crimes WHERE hour > 14', conn)# get and transform in a df

late_crimes = van_crimes_df.loc[van_crimes_df['HOUR'] > 18]# use a mask to get the crimes in hour after the 18 hours

dangerous_month_crimes = van_crimes_df.loc[:, 'MONTH'].value_counts().head(1)#get all the rows of the column month and count it

van_crimes_df
```

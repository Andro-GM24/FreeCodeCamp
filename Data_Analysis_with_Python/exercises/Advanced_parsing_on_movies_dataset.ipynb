```python
import numpy as np
import pandas as pd

column_names = ['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews',	'country',
                'cotent_rating', 'budget', 'title_year', 'imdb_score', 'genre']

dtype_dict={

            'budget':np.float64

}
```

```python
df=pd.read_csv('movies.csv', on_bad_lines='skip',sep='|',header=None,names=column_names,index_col=5,thousands=',', na_values=['?', 'NaN'])
#df[df=='?']
df.isna()
```

```python
def test_budget():
    assert df.loc[:, 'budget'].min() == 105000000.0

def test_country():
    assert '?' not in df.country.unique()

def test_shape():
    assert df.shape == (100, 12)
```

```python
test_budget()
test_country()
test_shape()
```
